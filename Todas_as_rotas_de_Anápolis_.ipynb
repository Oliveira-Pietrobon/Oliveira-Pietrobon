```python
#extração das coordenadas das rotas, e download do arquivo .csv com as coordenadas
import xml.etree.ElementTree as ET
import csv

def extract_coordinates_from_multiple_kml(kml_files, csv_file):
    # Abrir o arquivo CSV para escrita
    with open(csv_file, mode='w', newline='') as file:
        writer = csv.writer(file)

        # Iterar sobre todos os arquivos KML
        rota = 0
        for kml_file in kml_files:
            print(f"Processando arquivo: {kml_file}")

            # Parse o arquivo KML
            tree = ET.parse(kml_file)
            root = tree.getroot()

            # Namespaces para o KML
            namespaces = {'kml': 'http://www.opengis.net/kml/2.2'}

            # Procurar por todos os elementos de coordenadas no KML
            for placemark in root.findall('.//kml:Placemark', namespaces):
                for coordinates in placemark.findall('.//kml:coordinates', namespaces):
                    coord_list = coordinates.text.strip().split()
                    for coord in coord_list:
                        lon, lat, alt = coord.split(",")
                        # Escrever as coordenadas no arquivo CSV, incluindo o nome do arquivo
                        writer.writerow([rota, lat, lon, alt])
            rota += 1

    print(f"Coordenadas extraídas com sucesso para {csv_file}")

# Obter os nomes dos arquivos KML que foram carregados
kml_files = list(uploaded.keys())
csv_file = 'coordenadasRotasUrbanAnapolis.csv'

# Chamar a função para extrair as coordenadas de todos os arquivos KML
extract_coordinates_from_multiple_kml(kml_files, csv_file)

files.download(csv_file)
```

```python
dados = np.genfromtxt('coordenadasRotasUrbanAnapolis.csv', delimiter = ',')
print(dados)
```

```python
import folium
from folium import PolyLine
import pandas as pd
import numpy as np

# Criar o mapa base centrado em Anápolis
mapaAnapolis = folium.Map(location=[-16.3287, -48.95267], zoom_start=12)

folium.TileLayer(attr='FindBus', name='Mapa Anápolis').add_to(mapaAnapolis)

df = pd.DataFrame(dados, columns=['id', 'latitude', 'longitude', 'extra'])

# Agrupar os dados pelas rotas com base no ID
rotas = df.groupby('id')

contador_coordenadas = {}

# Função para atualizar a contagem de sobreposições
def contar_sobreposicao(coordenadas):
    for lat, lon in coordenadas:
        chave = (round(lat, 5), round(lon, 5))
        if chave not in contador_coordenadas:
            contador_coordenadas[chave] = 0
        contador_coordenadas[chave] += 1

for id_rota, pontos_rota in rotas:
    coordenadas = pontos_rota[['latitude', 'longitude']].values.tolist()
    contar_sobreposicao(coordenadas)

# Iterar sobre cada rota e criar uma linha no mapa
for id_rota, pontos_rota in rotas:
    coordenadas = pontos_rota[['latitude', 'longitude']].values.tolist()

    opacidade = np.mean([contador_coordenadas[(round(lat, 5), round(lon, 5))] for lat, lon in coordenadas])
    opacidade_normalizada = min(1, max(0.3, opacidade / 10))
    PolyLine(coordenadas, color="blue", weight=2.5, opacity=opacidade_normalizada).add_to(mapaAnapolis)

folium.LayerControl(position="topleft").add_to(mapaAnapolis)

mapaAnapolis
```
